```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(-100, 100)
sigmoid = lambda x : 1 / (1 + np.exp(-x))
plt.figure(dpi=120)
plt.plot(x, sigmoid(x), label="sigmoid")
plt.plot(x, sigmoid(x) * (1 - sigmoid(x)), label="sigmoid derivatives", ls="--")
plt.legend()
plt.show()
```
